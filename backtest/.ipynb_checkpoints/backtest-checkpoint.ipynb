# pct_ladder Strategy Backtester
Configure parameters below, then **Run All** (`Kernel → Restart & Run All`).

```python
import sys, os
sys.path.insert(0, os.path.dirname(os.path.abspath('.')))  # project root

from decimal import Decimal
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
print('Imports OK')
```

## ⚙️ Configuration
Edit the cells below to change parameters.

```python
# ── Date range ────────────────────────────────────────────────────────────────
START_DATE  = '2022-01-01'
END_DATE    = '2026-01-01'
INTERVAL    = '1m'          # 1m | 5m | 15m | 1h | 4h | 1d

# ── Capital ──────────────────────────────────────────────────────────────────
INITIAL_CASH      = 25_000   # starting USDC balance
INITIAL_ETH       = 5.0      # starting ETH holding
INITIAL_ETH_COST  = 3_000    # avg cost basis of initial ETH (0 = unknown)

# ── Strategy ─────────────────────────────────────────────────────────────────
UPPER_PCT   = 0.4     # sell threshold %
LOWER_PCT   = 0.4     # buy threshold %
BUY_QUOTE   = 1400    # USDC per buy step
SELL_QUOTE  = 1400    # USDC per sell step
QTY_STEP    = 1e-6
MIN_QTY     = 1e-6

# ── Rebalance ────────────────────────────────────────────────────────────────
REBALANCE_THRESHOLD_STEPS = 4   # trigger rebalance when side has ≤ N steps
REBALANCE_TARGET_STEPS    = 8   # restore to N steps after rebalance

# ── Execution ────────────────────────────────────────────────────────────────
QUOTE_RESERVE       = 500    # keep this much USDC as reserve (never buy with it)
USE_INVENTORY_BUFFER = True  # allow selling initial ETH holding

# ── Data source ──────────────────────────────────────────────────────────────
SYMBOL       = 'ETHUSDC'    # trading symbol
FETCH_SYMBOL = 'ETHUSDT'    # data source symbol (Binance; USDC≈USDT)
DATA_SOURCE  = 'binance'    # 'binance' has 4yr history; 'mexc' only recent
CACHE_DIR    = 'cache'
```

## 📥 Fetch Historical Data

```python
from backtest.fetch_data import fetch_klines, dt_to_ms

start_ms = dt_to_ms(START_DATE)
end_ms   = dt_to_ms(END_DATE)

df_candles = fetch_klines(
    symbol    = FETCH_SYMBOL,
    interval  = INTERVAL,
    start_ms  = start_ms,
    end_ms    = end_ms,
    cache_dir = CACHE_DIR,
    source    = DATA_SOURCE,
)

print(f'Loaded {len(df_candles):,} candles  |  {df_candles["ts"].iloc[0]}  →  {df_candles["ts"].iloc[-1]}')
df_candles.tail(3)
```

## 🔁 Run Backtest

```python
from backtest.engine import BacktestEngine
from strategies.pct_ladder.strategy import create_strategy

strategy_cfg = dict(
    symbols                   = [SYMBOL],
    upper_pct                 = UPPER_PCT,
    lower_pct                 = LOWER_PCT,
    sizing_mode               = 'fixed_quote',
    buy_quote                 = BUY_QUOTE,
    sell_quote                = SELL_QUOTE,
    qty_step                  = QTY_STEP,
    min_qty                   = MIN_QTY,
    rebalance_threshold_steps = REBALANCE_THRESHOLD_STEPS,
    rebalance_target_steps    = REBALANCE_TARGET_STEPS,
)

strategy = create_strategy(strategy_cfg)

engine = BacktestEngine(
    strategy            = strategy,
    symbol              = SYMBOL,
    initial_cash        = Decimal(str(INITIAL_CASH)),
    initial_eth         = Decimal(str(INITIAL_ETH)),
    initial_eth_cost    = Decimal(str(INITIAL_ETH_COST)),
    quote_reserve       = Decimal(str(QUOTE_RESERVE)),
    use_inventory_buffer= USE_INVENTORY_BUFFER,
)

candles = df_candles.to_dict('records')
print(f'Running backtest on {len(candles):,} candles...')
summary = engine.run(candles)
print('Done!')
```

## 📊 Summary

```python
days  = (pd.Timestamp(END_DATE) - pd.Timestamp(START_DATE)).days
years = days / 365

init_eq  = summary['initial_equity']
final_eq = summary['final_equity']
pnl_abs  = final_eq - init_eq
pnl_pct  = pnl_abs / init_eq * 100 if init_eq else 0
ann_ret  = ((final_eq / init_eq) ** (1 / years) - 1) * 100 if init_eq and years > 0 else 0
cpd      = summary['cycles'] / days if days else 0
avg_cyc  = summary['realized_pnl'] / summary['cycles'] if summary['cycles'] else 0

rows = [
    ('Period',                f"{START_DATE}  →  {END_DATE}  ({days:,} days)"),
    ('Initial equity',        f"${init_eq:,.2f}"),
    ('Final equity',          f"${final_eq:,.2f}"),
    ('Total PnL',             f"${pnl_abs:,.2f}  ({pnl_pct:.1f}%)"),
    ('Realized PnL (ladder)', f"${summary['realized_pnl']:,.2f}"),
    ('Annualised return',     f"{ann_ret:.1f}%"),
    ('Max drawdown',          f"{summary['max_drawdown_pct']:.2f}%"),
    ('Total cycles',          f"{summary['cycles']:,}"),
    ('Cycles / day',          f"{cpd:.2f}"),
    ('Avg PnL / cycle',       f"${avg_cyc:.2f}"),
    ('Ladder buys',           f"{summary['ladder_buys']:,}"),
    ('Ladder sells',          f"{summary['ladder_sells']:,}"),
    ('Rebalance trades',      f"{summary['rebalance_trades']:,}"),
]

df_summary = pd.DataFrame(rows, columns=['Metric', 'Value'])
df_summary.style.set_properties(**{'text-align': 'left'}).hide(axis='index')
```

## 📅 Monthly & Yearly Breakdown

```python
df_trades = pd.DataFrame(engine.trades)
df_trades['ts'] = pd.to_datetime(df_trades['ts'], utc=True, errors='coerce')
df_trades['month'] = df_trades['ts'].dt.to_period('M')
df_trades['year']  = df_trades['ts'].dt.year

ladder = df_trades[~df_trades['reason'].str.startswith('rebalance_')]

# Monthly aggregation
m_buys  = ladder[ladder['side']=='BUY'].groupby('month').size().rename('buys')
m_sells = ladder[ladder['side']=='SELL'].groupby('month').size().rename('sells')

sells_sorted = ladder[ladder['side']=='SELL'].sort_values('ts')
last_pnl     = sells_sorted.groupby('month')['realized_pnl_cumulative'].last()
m_pnl        = last_pnl.diff().fillna(last_pnl.iloc[0]).rename('realized_pnl').round(2)

df_monthly = pd.concat([m_buys, m_sells, m_pnl], axis=1).fillna(0)
df_monthly['cycles'] = df_monthly[['buys','sells']].min(axis=1).astype(int)
df_monthly['buys']   = df_monthly['buys'].astype(int)
df_monthly['sells']  = df_monthly['sells'].astype(int)
df_monthly = df_monthly.reset_index()
df_monthly['month'] = df_monthly['month'].astype(str)
df_monthly = df_monthly[['month','buys','sells','cycles','realized_pnl']]

# Yearly aggregation
df_yearly = df_monthly.copy()
df_yearly['year'] = df_yearly['month'].str[:4]
df_yearly = (df_yearly.groupby('year')
    .agg(buys=('buys','sum'), sells=('sells','sum'),
         cycles=('cycles','sum'), realized_pnl=('realized_pnl','sum'))
    .reset_index()
    .assign(realized_pnl=lambda d: d['realized_pnl'].round(2)))

print('=== Yearly Summary ===')
display(df_yearly.style.format({'realized_pnl': '${:,.2f}'}).hide(axis='index'))
print('\n=== Monthly Breakdown ===')
df_monthly.style.format({'realized_pnl': '${:,.2f}'}).hide(axis='index')
```

## 📈 Equity Curve

```python
eq_ts  = pd.to_datetime([t for t, _ in engine.equity_curve], utc=True)
eq_val = [v for _, v in engine.equity_curve]

# Downsample to hourly for plotting speed
df_eq = pd.DataFrame({'ts': eq_ts, 'equity': eq_val}).set_index('ts')
df_eq = df_eq.resample('1h').last().dropna()

# ETH price (from candles, resampled)
df_px = df_candles.set_index('ts')['close'].resample('1h').last().dropna()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True,
                                gridspec_kw={'height_ratios': [2, 1]})

ax1.plot(df_eq.index, df_eq['equity'], color='#2196F3', linewidth=1.2, label='Portfolio equity')
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax1.set_ylabel('Equity (USD)')
ax1.set_title(f'Equity Curve  |  {START_DATE} → {END_DATE}  |  {UPPER_PCT}% / ${BUY_QUOTE}')
ax1.legend()

ax2.plot(df_px.index, df_px.values, color='#FF9800', linewidth=1, label='ETH price')
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax2.set_ylabel('ETH Price (USD)')
ax2.set_xlabel('Date')
ax2.legend()

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

## 💰 Monthly Realized PnL

```python
fig, ax = plt.subplots(figsize=(14, 5))

colors = ['#4CAF50' if v >= 0 else '#F44336' for v in df_monthly['realized_pnl']]
bars = ax.bar(df_monthly['month'], df_monthly['realized_pnl'], color=colors, width=0.7)

ax.axhline(0, color='black', linewidth=0.8)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.set_xlabel('Month')
ax.set_ylabel('Realized PnL (USD)')
ax.set_title('Monthly Realized PnL  (ladder trades only)')
plt.xticks(rotation=90, fontsize=7)
plt.tight_layout()
plt.show()
```

## 📆 Monthly Cycles

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.bar(df_monthly['month'], df_monthly['cycles'], color='#9C27B0', width=0.7)
ax.set_xlabel('Month')
ax.set_ylabel('Cycles')
ax.set_title('Monthly Cycles (complete round trips)')
plt.xticks(rotation=90, fontsize=7)
plt.tight_layout()
plt.show()
```

## 📈 Cumulative Realized PnL

```python
df_cum = df_monthly[['month','realized_pnl']].copy()
df_cum['cumulative_pnl'] = df_cum['realized_pnl'].cumsum()

fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(df_cum['month'], df_cum['cumulative_pnl'], marker='o', markersize=3,
        color='#2196F3', linewidth=1.5)
ax.fill_between(df_cum['month'], df_cum['cumulative_pnl'], alpha=0.15, color='#2196F3')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.set_xlabel('Month')
ax.set_ylabel('Cumulative Realized PnL (USD)')
ax.set_title('Cumulative Realized PnL')
plt.xticks(rotation=90, fontsize=7)
plt.tight_layout()
plt.show()
```

## 📉 Drawdown

```python
eq_series = df_eq['equity']
rolling_peak = eq_series.cummax()
drawdown_pct = (eq_series - rolling_peak) / rolling_peak * 100

fig, ax = plt.subplots(figsize=(14, 4))
ax.fill_between(drawdown_pct.index, drawdown_pct.values, 0, color='#F44336', alpha=0.5)
ax.plot(drawdown_pct.index, drawdown_pct.values, color='#F44336', linewidth=0.8)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.1f}%'))
ax.set_xlabel('Date')
ax.set_ylabel('Drawdown')
ax.set_title(f'Drawdown  |  Max: {summary["max_drawdown_pct"]:.2f}%')
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

## 🔄 Config Comparison
Compare multiple `(upper_pct, buy_quote)` configs side by side.

```python
# Define configs to compare — add/remove as needed
COMPARE_CONFIGS = [
    dict(label='0.4% / $1400',  upper_pct=0.4, lower_pct=0.4, buy_quote=1400, sell_quote=1400),
    dict(label='0.6% / $1500',  upper_pct=0.6, lower_pct=0.6, buy_quote=1500, sell_quote=1500),
    dict(label='0.5% / $1000',  upper_pct=0.5, lower_pct=0.5, buy_quote=1000, sell_quote=1000),
    dict(label='1.0% / $2000',  upper_pct=1.0, lower_pct=1.0, buy_quote=2000, sell_quote=2000),
]

compare_results = []

for cfg in COMPARE_CONFIGS:
    scfg = dict(
        symbols                   = [SYMBOL],
        sizing_mode               = 'fixed_quote',
        qty_step                  = QTY_STEP,
        min_qty                   = MIN_QTY,
        rebalance_threshold_steps = REBALANCE_THRESHOLD_STEPS,
        rebalance_target_steps    = REBALANCE_TARGET_STEPS,
        **{k: v for k, v in cfg.items() if k != 'label'},
    )
    strat = create_strategy(scfg)
    eng   = BacktestEngine(
        strategy            = strat,
        symbol              = SYMBOL,
        initial_cash        = Decimal(str(INITIAL_CASH)),
        initial_eth         = Decimal(str(INITIAL_ETH)),
        initial_eth_cost    = Decimal(str(INITIAL_ETH_COST)),
        quote_reserve       = Decimal(str(QUOTE_RESERVE)),
        use_inventory_buffer= USE_INVENTORY_BUFFER,
    )
    s = eng.run(candles)
    d = days
    compare_results.append({
        'Config':             cfg['label'],
        'Realized PnL':       round(s['realized_pnl'], 2),
        'Final Equity':       round(s['final_equity'], 2),
        'Total PnL%':         round((s['final_equity'] - s['initial_equity']) / s['initial_equity'] * 100, 1),
        'Annualised%':        round(((s['final_equity'] / s['initial_equity']) ** (365/d) - 1) * 100, 1) if d else 0,
        'Cycles':             s['cycles'],
        'Cycles/Day':         round(s['cycles'] / d, 2) if d else 0,
        'Avg PnL/Cycle':      round(s['realized_pnl'] / s['cycles'], 2) if s['cycles'] else 0,
        'Max Drawdown%':      s['max_drawdown_pct'],
        'Rebalance Trades':   s['rebalance_trades'],
    })
    print(f"  {cfg['label']:20s}  cycles={s['cycles']:,}  pnl=${s['realized_pnl']:,.0f}")

df_compare = pd.DataFrame(compare_results)
df_compare.style\
    .format({'Realized PnL': '${:,.2f}', 'Final Equity': '${:,.2f}',
             'Total PnL%': '{:.1f}%', 'Annualised%': '{:.1f}%',
             'Avg PnL/Cycle': '${:.2f}', 'Max Drawdown%': '{:.2f}%'})\
    .highlight_max(subset=['Realized PnL','Cycles','Annualised%'], color='#c8e6c9')\
    .highlight_min(subset=['Max Drawdown%'], color='#c8e6c9')\
    .hide(axis='index')
```

```python
# Comparison bar charts
labels = df_compare['Config']
x = range(len(labels))

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

axes[0].bar(labels, df_compare['Realized PnL'], color='#4CAF50')
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'${v:,.0f}'))
axes[0].set_title('Realized PnL')
axes[0].tick_params(axis='x', rotation=15)

axes[1].bar(labels, df_compare['Cycles'], color='#9C27B0')
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'{v:,.0f}'))
axes[1].set_title('Total Cycles')
axes[1].tick_params(axis='x', rotation=15)

axes[2].bar(labels, df_compare['Annualised%'], color='#2196F3')
axes[2].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'{v:.0f}%'))
axes[2].set_title('Annualised Return')
axes[2].tick_params(axis='x', rotation=15)

plt.suptitle(f'Config Comparison  |  {START_DATE} → {END_DATE}', fontsize=13)
plt.tight_layout()
plt.show()
```
